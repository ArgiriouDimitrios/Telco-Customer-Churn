```python
sns.countplot(x="Churn",data=churn)
plt.title("ChurnDistribution")
plt.show()
```

```python
categorical_cols = churn.select_dtypes(include='object').columns.tolist()
categorical_cols.remove('customerID')  # Exclude ID
categorical_cols.remove('Churn')       # Exclude target

for col in categorical_cols:
    plt.figure(figsize=(6,4))
    sns.countplot(y=col, data=churn)
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
numerical_cols = churn.select_dtypes(include=['int64','float64']).columns.tolist()

for col in numerical_cols:
    plt.figure(figsize=(6,4))
    sns.histplot(churn[col], kde=True, bins=30)
    plt.title(f"Distribution of {col}")
    plt.show()
```
